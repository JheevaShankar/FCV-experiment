```python
import cv2
from google.colab.patches import cv2_imshow

# Read your image with new path
img = cv2.imread("/content/Endo Yamato.jpeg")

if img is None:
    print(" ERROR: Image not loaded. Check the path or file format.")
else:
    # Resize to a smaller size
    img_small = cv2.resize(img, (256, 256))  # or change to (128, 128) for even smaller

    # Show the resized image
    cv2_imshow(img_small)

    # Optionally, save the resized image
    cv2.imwrite("/content/Endo Yamato.jpeg", img_small)
    print(" Saved resized image as /content/Endo Yamato.jpeg")
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Endo Yamato.jpeg")
cv2_imshow(img)
print("Shape:", img.shape)

img = cv2.imread("/content/Endo Yamato.jpeg")
if img is None:
    print("ERROR: UNABLE TO READ IMAGE.")
else:
    print("IMAGE LOADED SUCCESSFULLY")
    print("Shape:", img.shape)

img_crop = img[0:200, 0:200]
cv2_imshow(img_crop)

cv2.imwrite("/content/Endo Yamato.jpeg", img_crop)
print("Cropped image saved as /content/Endo Yamato.jpeg")

img_resize = cv2.resize(img, (256, 256))
cv2_imshow(img_resize)
print("Resized shape:", img_resize.shape)

img_resize_half = cv2.resize(img, (img.shape[1] // 2, img.shape[0] // 2))
cv2_imshow(img_resize_half)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, th1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img_gray, 200, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

desired_size = (200, 200)
th1_resized = cv2.resize(th1, desired_size)
th2_resized = cv2.resize(th2, desired_size)
th3_resized = cv2.resize(th3, desired_size)
th4_resized = cv2.resize(th4, desired_size)
th5_resized = cv2.resize(th5, desired_size)
gray_resized = cv2.resize(img_gray, desired_size)

combined = np.hstack((gray_resized, th1_resized, th2_resized, th3_resized, th4_resized, th5_resized))

cv2_imshow(combined)

cv2.destroyAllWindows()
```
